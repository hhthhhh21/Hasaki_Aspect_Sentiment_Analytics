## Phân tích nhãn thực thể và nhãn cảm xúc

```python
import google.generativeai as genai
import os
import pandas as pd
import time

# Hàm phân tích và gắn nhãn dữ liệu
def label_analysis(text, model):
    # Viết yêu cầu
    request = '''Hãy gắn nhãn dữ liệu cho các bình luận phản hồi của khách hàng sau khi mua mỹ phẩm theo yêu cầu dưới đây. Hãy tách câu, sao cho mỗi câu được tách ra từ INPUT sẽ chỉ được thuộc về một nhãn thực thể và một nhãn cảm xúc, bắt buộc phải có cả nhãn thực thể và nhãn cảm xúc. Chỉ sử dụng nhãn ở dưới đây, không thêm bất kỳ nhãn nào khác. Hãy viết OUTPUT như ví dụ phía dưới. Lưu ý không thêm bất kỳ chú thích nào khác.

                    - Nhãn cảm xúc gồm 3 nhãn sau:
                        + positive: tức nhận xét tích cực, khen. Các nhận xét có các từ như: tốt, đẹp, được, sạch sẽ, ok, thoáng, mát, tuyệt vời, xịn, mịn, thân thiện, vui vẻ, lịch sự, chuyên nghiệp, cẩn thận, tuyệt, mịn màng, tương đối, đáng tiền, hấp dẫn, ổn, hợp lý, rất hợp lý, phải chăng, phù hợp, khá mềm, vừa phải, mát, mát lạnh,...
                        + negative: tức nhận xét tiêu cực, chê. Các nhận xét có các từ như: xấu, tồi, tệ, kém, bẩn, chưa được tốt, bí, ẩm thấp, mụn, không đầy đủ, cũ, không sang trọng, xa, ngõ hẻm, không chuyên nghiệp, thiếu chuyên nghiệp, không vui vẻ, dở, không ngon, không hợp khẩu vị, mặn, nhạt, đắt, quá đắt, cao, quá cao, hơi cao,...
                        + neutral: tức nằm giữa tích cực và tiêu cực (trung dung). Các nhận xét có các từ như: tạm được, bình thường, không có gì đặc sắc, chấp nhận được, tạm ổn, không quá đắt, vừa phải,...

                    - Nhãn thực thể gồm 6 nhãn sau:
                        + PACKAGING: đề cập đến bao bì của sản phẩm như hộp, túi, cách đóng gói, seal, nhãn mác.
                        + QUALITY: đề cập đến chất lượng sản phẩm như hiệu quả, độ bền, độ an toàn, thành phần, hương thơm.
                        + PRICE: đề cập tới giá cả của sản phẩm, dịch vụ cung cấp.
                        + SERVICE: đề cập tới thái độ phục vụ, hỗ trợ khách hàng, thời gian giao hàng.
                        + STORE: đề cập tới cửa hàng, quầy mỹ phẩm hoặc nền tảng thương mại điện tử bán sản phẩm.
                        + MISCELLANEOUS: cho các câu bình luận không nằm trong những thực thể ở trên.

                    - Ví dụ:
                        INPUT: Sản phẩm đóng gói rất đẹp. Hiệu quả dưỡng ẩm rất tốt. Giá cả hợp lý. Giao hàng chậm, cần cải thiện. Bao bì chắc chắn, nhưng sản phẩm hơi đắt.
                        OUTPUT: Sản phẩm đóng gói rất đẹp ==> PACKAGING: positive
                                Hiệu quả dưỡng ẩm rất tốt ==> QUALITY: positive
                                Giá cả hợp lý ==> PRICE: positive
                                Giao hàng chậm, cần cải thiện ==> SERVICE: negative
                                Bao bì chắc chắn ==> PACKAGING: positive
                                nhưng sản phẩm hơi đắt ==> PRICE: negative

                    - Lưu ý chỉ in dòng này thôi:
                      "Sản phẩm đóng gói rất đẹp ==> PACKAGING: positive
                      Hiệu quả dưỡng ẩm rất tốt ==> QUALITY: positive
                      Giá cả hợp lý ==> PRICE: positive
                      Giao hàng chậm, cần cải thiện ==> SERVICE: negative
                      Bao bì chắc chắn ==> PACKAGING: positive
                      nhưng sản phẩm hơi đắt ==> PRICE: negative"
                    '''
    prompt = f"{request}\n{text}"

    # Retry logic in case of errors
    retries = 3
    for attempt in range(retries):
        try:
            # Gửi yêu cầu đến model
            response = model.generate_content(prompt)
            return response.text.strip()
        except Exception as err:
            print(f"Attempt {attempt + 1} failed: {err}")
            time.sleep(2)  # Wait before retrying
    # Trả về giá trị ban đầu nếu không thành công
    return text

# Cấu hình sinh văn bản
generation_config = {
    'candidate_count': 1,
    'max_output_tokens': 4096,
    'temperature': 1.0,
    'top_p': 0.5,
}

# Cài đặt an toàn
safety_settings = [
    {'category': 'HARM_CATEGORY_DANGEROUS', 'threshold': 'BLOCK_NONE'},
    {'category': 'HARM_CATEGORY_HARASSMENT', 'threshold': 'BLOCK_NONE'},
    {'category': 'HARM_CATEGORY_HATE_SPEECH', 'threshold': 'BLOCK_NONE'},
    {'category': 'HARM_CATEGORY_SEXUALLY_EXPLICIT', 'threshold': 'BLOCK_NONE'},
    {'category': 'HARM_CATEGORY_DANGEROUS_CONTENT', 'threshold': 'BLOCK_NONE'},
]

# Thiết lập API key từ biến môi trường
genai.configure(api_key=os.environ.get('GOOGLE_API_KEY', 'api_key_hehe'))

# Kiểm tra nếu cột 'Content' có tồn tại trong file dữ liệu
input_file_path = r"C:\Users\84913\OneDrive\Documents\Bon.xlsx"# Update with your input file path
if os.path.exists(input_file_path):
    data = pd.read_excel(input_file_path)
    if 'Content' in data.columns:
        # Khởi tạo model
        model = genai.GenerativeModel('gemini-pro', generation_config=generation_config, safety_settings=safety_settings)

        # Gắn nhãn từng bình luận
        data['Label analysis'] = data['Content'].map(lambda x: label_analysis(x, model))

        # Lưu kết quả ra file
        output_file_path = 'cleaned_data.csv'
        data.to_csv(output_file_path, encoding='utf-8-sig', index=False)

        # Hiển thị một số dòng đầu tiên
        print(data.head())
    else:
        print("'Content' column not found in the input file!")
else:
    print(f"Input file '{input_file_path}' not found!")
```

## Bóc tách nhãn thực thể và nhãn cảm xúc

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
data=pd.read_csv(r"C:\Users\84913\Downloads\cleaned_data.csv",encoding='utf-8-sig')
data['content_ID'] = [str(i).zfill(4) for i in range(len(data))]
data
```

```python
def create_data(column):
    entity_label = ['PACKAGING', 'QUALITY', 'PRICE', 'SERVICE', 'STORE', 'MISCELLANEOUS']
    sentiment_label = ['positive', 'negative', 'neutral']
    list_newdata = []
    for i in range(0, len(column)):
        line_breaks = [0] + [index+1 for index, char in enumerate(column.iloc[i]) if char == '\n']
        for j in range(0, len(line_breaks)):
            x = column.iloc[i].find('==> ', line_breaks[j]) + len('==> ')
            if line_breaks[j] == line_breaks[-1]:
                label = column.iloc[i][x:].strip()
            else:
                y = line_breaks[j+1]
                label = column.iloc[i][x:y].strip()
            kt = label.split(': ')
            if (len(label) >= 13) and (len(label) <= 24) and (len(kt) == 2) and (kt[0] in entity_label) and (kt[1] in sentiment_label) and (len(kt[0]) <= 14) and (len(kt[0]) >= 4) and (len(kt[1]) <= 8) and (len(kt[1]) >= 7):
                dict_newdata = {}
                dict_newdata['content_ID'] = data['content_ID'].iloc[i]
                x0 = line_breaks[j]
                y0 = column.iloc[i].find(' ==>', x0)
                dict_newdata['content'] = column.iloc[i][x0:y0].strip()
                x1 = y0 + len(' ==> ')
                y1 = column.iloc[i].find(':', x1)
                dict_newdata['Entity label'] = column.iloc[i][x1:y1].strip()
                x2 = y1 + 2
                if line_breaks[j] == line_breaks[-1]:
                    y2 = len(column.iloc[i])
                else:
                    y2 = line_breaks[j+1] - 1
                dict_newdata['Sentiment label'] = column.iloc[i][x2:y2].strip()
                list_newdata.append(dict_newdata)
    newdata = pd.DataFrame(list_newdata)
    return newdata
model_data = create_data(data['Label analysis'])
print(model_data)
```

## Gộp commment và các aspect

```python
import pandas as pd

# Đọc file Excel
file_path = r"C:\Users\84913\Downloads\bon.csv"
df = pd.read_csv(file_path,encoding='utf-8-sig')

# Lấy danh sách các aspect duy nhất
aspects = df['Entity label'].unique()

# Tạo DataFrame mới với ID, comment, và mỗi aspect là một cột
result = pd.DataFrame(columns=['content_ID', 'content'] + list(aspects))
result['content_ID'] = df['content_ID'].unique()  # Thêm các giá trị ID duy nhất

# Khởi tạo tất cả các cột aspect với giá trị "neutral"
for aspect in aspects:
    result[aspect] = "neutral"

# Gộp comment và xác định sentiment
for current_id in result['content_ID']:
    # Gộp comment cho từng ID
    comments = df.loc[df['content_ID'] == current_id, 'content'].unique()
    comments = [str(comment) for comment in comments]  # Chuyển tất cả phần tử thành chuỗi
    result.loc[result['content_ID'] == current_id, 'content'] = " ".join(comments)  # Gộp các comment lại

    # Điền sentiment cho các aspect
    rows = df[df['content_ID'] == current_id]
    for aspect in aspects:
        # Lấy tất cả sentiment liên quan đến aspect hiện tại
        sentiments = rows.loc[rows['Entity label'] == aspect, 'Sentiment label']
        if not sentiments.empty:
            # Đếm số lượng mỗi sentiment
            sentiment_counts = sentiments.value_counts()
            # Chọn sentiment có số lượng lớn nhất
            majority_sentiment = sentiment_counts.idxmax()
            result.loc[result['content_ID'] == current_id, aspect] = majority_sentiment

# Xuất kết quả ra file Excel mới
output_file = r"Data_KN_Processed.xlsx"
result.to_excel(output_file, index=False, engine='openpyxl')
print(f"Kết quả đã được lưu vào: {output_file}")
```
